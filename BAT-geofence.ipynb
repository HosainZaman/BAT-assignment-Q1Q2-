```python
geo_df['Longitude']=geo_df['Longitude'].round(decimals=4)
geo_df['Latitude']=geo_df['Latitude'].round(decimals=5)
```

```python
ret_loc['long']=ret_loc['long'].round(decimals=4)
ret_loc['lat']=ret_loc['lat'].round(decimals=5)
```

```python
geo_df=geo_df[geo_df['Longitude'].isin(ret_loc['long']) & geo_df['Latitude'].isin(ret_loc['lat'])]
```

```python
x1,y1,x2,y2 = geo_df['geometry'].total_bounds
```

```python
geo_df.crs = "epsg:4326"
geo_df.to_crs = {'init' :'epsg:4326'}
```

```python
m = folium.Map(tiles='openstreetmap')
m.fit_bounds([[y1, x1], [y2, x2]])
folium.GeoJson(geo_df["geometry"]).add_to(m)
```
